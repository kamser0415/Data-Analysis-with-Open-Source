```python
import requests
import json
import pandas as pd

def get_year_months():
  year_months = []
  for year in range(2015, 2024+1):
    for month in range(1,12+1):
      year_months.append(f'{year}{month:02d}')
  return year_months

data = []

url = "http://apis.data.go.kr/1613000/ApHusEnergyUseInfoOfferServiceV2/getWntyAvrgEnergyUseAmountInfoSearchV2"
serviceKey = "0f6bbcacb363a3809c03696ba2df52dddcf07981a315ca5d5c4092f1aaeafefc"

try :
  for month in get_year_months() :
    params = {
        'serviceKey': serviceKey,
        'searchDate': month
    }
    response = requests.get(url,params = params)
    response.raise_for_status()

    # 데이터 넣기
    item = response.json()['response']['body']['item']
    item['date_str'] = month
    data.append(item)

  print('API 호출 완료')
except Exception as e:
  print(f"오류발생 {e}")

with open("energy_data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("저장 완료: energy_data.json")
```

```python
# 2-2 연도별, 계절별 분석을 위해 날짜 컬럼을 활용하여 ‘year’와 ‘season’ 컬럼을 추가하는 전처리
# 코드를 작성하고, 변환 결과를 확인할 수 있는 출력 결과를 첨부하시오.

df['date'] = pd.to_datetime(df['date_str'],format='%Y%m')
df['year'] = df['date'].dt.year

df["season"] = df["date"].dt.month.map(
    lambda m: "봄" if 3 <= m <= 5
    else "여름" if 6 <= m <= 8
    else "가을" if 9 <= m <= 11
    else "겨울"
)

df
```
